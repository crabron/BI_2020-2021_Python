Картинка распределения длин последовательностей

```python
import seaborn as sns
import matplotlib.pyplot as plt
from Bio import SeqIO

#path to fastq file
path="/home/gladkov/storage/temp/temp_fast.fastq"

sizes = [len(rec) for rec in SeqIO.parse(path, 'fastq')]
plt.figure(figsize=(16, 6))

sns.distplot(sizes, hist=False, kde=True, 
             color = 'salmon', 
             kde_kws={'linewidth': 2})
```

Просто картинка на основе случайных данных

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(190121)

x = np.random.normal(1, 100, 100)
y = np.random.negative_binomial(100, 0.8, 100) * x * np.random.randint(0, 10, 100)
z = ["left_char"]*50 + ["right_char"]*50

df = pd.DataFrame({'normal':x,'some_neg_binomial_with_noise':y, 'char':z})

sns.set_style("whitegrid")
sns.lmplot(data=df,
    x="normal",
    y="some_neg_binomial_with_noise",
    hue="char",
    palette="viridis",
    height=10
)
```
